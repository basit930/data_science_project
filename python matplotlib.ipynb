```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,43,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days,max_t)
```

```python
plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")
plt.xlabel("days")
plt.ylabel("temperature")
plt.title("weather")
plt.legend()
```

# you can change location of the legend

```python
plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")
plt.xlabel("days")
plt.ylabel("temperature")
plt.title("weather")
plt.legend(loc="upper right")#you can perform these function 1:upper left,2:lower left,3:center,4:center right or left
```

```python
#best method is good 
plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")
plt.xlabel("days")
plt.ylabel("temperature")
plt.title("weather")
plt.legend(loc="best",shadow=True,fontsize="large")
```

```python
#best method is good 
plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")
plt.xlabel("days")
plt.ylabel("temperature")
plt.title("weather")
plt.legend(loc="best",shadow=True,fontsize="large")
plt.grid()
```

```python
import numpy as np
```

```python
company=["google","azn","msft","fb"]
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
ypos=np.arange(len(company))
ypos
```

```python
plt.bar(company,revenue)
```

```python
plt.bar(ypos,revenue)
```

```python
# you can change the number of x axis to the name this function use xticks
plt.xticks(ypos,company)
plt.bar(ypos,revenue)
```

```python
plt.xticks(ypos,company)
plt.ylabel("revenue(bin)")
plt.title("us tech stacks")
plt.bar(ypos,revenue,label="revenue")
plt.legend()
```

# this method we can calculate the profit 

```python
plt.xticks(ypos,company)
plt.ylabel("revenue(bin)")
plt.title("us tech stacks")
plt.bar(ypos,revenue,label="revenue")
plt.bar(ypos,profit,label="profit")
plt.legend()
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.xticks(ypos,company)
plt.ylabel("revenue(bin)")
plt.title("us tech stacks")
plt.bar(xpos-0.2,revenue,label="revenue")
plt.bar(xpos+0.2,profit,label="profit")
plt.legend()
```

```python
plt.xticks(ypos,company)
plt.ylabel("revenue(bin)")
plt.title("us tech stacks")
plt.bar(xpos-0.2,revenue,width=0.4,label="revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="profit")
plt.legend()
```

# plotting graph in horizental way

```python
plt.yticks(ypos,company)
plt.title("us tech stacks")
plt.barh(xpos-0.2,revenue,label="revenue")
plt.barh(xpos+0.2,profit,label="profit")
plt.legend()
```

```python
blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]
```

```python
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=["green","orange"],
        label=["men","women"])
plt.legend()
plt.xlabel("sugar range")
plt.ylabel("total no patients")
plt.title("blood sugar analysis")
```

```python
#horizental plotting
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=["green","orange"],
        label=["men","women"],orientation="horizontal")
plt.legend()
plt.xlabel("sugar range")
plt.ylabel("total no patients")
plt.title("blood sugar analysis")
```

# pie chart

```python
exp_vals=[1400,600,300,410,250]
exp_labels=["home rent","food","phone/internet bill","car","other utilities"]
```

```python
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=3)
plt.show()
```

```python
#this method is used to define percentage
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2,autopct="%0.2f%%")
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2,autopct="%0.1f%%")
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2,autopct="%0.3f%%")#0.3 means how many digit to get after the ratio
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2,autopct="%0.1f%%",shadow=True)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2,autopct="%0.1f%%",shadow=True,explode=[0,0,0,0,0])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2,autopct="%0.1f%%",shadow=True,explode=[0,0.2,0,0,0])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2,autopct="%0.1f%%",shadow=True,explode=[0,0.5,0,0,0])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2,autopct="%0.1f%%",shadow=True,explode=[0,0.2,0.2,0,0],startangle=45)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2,autopct="%0.1f%%",shadow=True,explode=[0,0.2,0.2,0,0],startangle=180)
plt.show()
```

# how to save the  chart in the file

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2,autopct="%0.1f%%",shadow=True,explode=[0,0.2,0.2,0,0],startangle=45)
plt.savefig("piechart.png")
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2,autopct="%0.1f%%",shadow=True,explode=[0,0.2,0.2,0,0],startangle=45)
plt.savefig("piechart2.png",bbox_inches="tight",pad_inches=2,transparent=True)
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2,autopct="%0.1f%%",shadow=True,explode=[0,0.2,0.2,0,0],startangle=45)
plt.savefig("piechart2.pdf",bbox_inches="tight",pad_inches=2,transparent=True)
```
