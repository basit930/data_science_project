# heatmap

```python
flights=sns.load_dataset("flights")
```

```python
flights.head()
```

```python
#first we create the pivot table to show to heatmap
x=flights.pivot_table(index="year",columns="month",values="passengers",aggfunc="sum")
```

```python
x
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x,cbar=False)
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5)
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5,annot=True,fmt="d")
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5,cmap="viridis",annot=True,fmt="d")
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5,cmap="summer",annot=True,fmt="d")
```

# cluster map

```python
sns.clustermap(x)
```

```python
sns.clustermap(x,metric="correlation")
```

```python
sns.clustermap(x,metric="correlation",z_score=0)
```

```python
sns.clustermap(x,metric="correlation",z_score=0,annot=True)
```

```python
sns.clustermap(x,metric="correlation",z_score=0,row_cluster=False,annot=True)
```

```python
sns.clustermap(x,metric="correlation",z_score=0,col_cluster=False,annot=True)
```

# distribution or dist  plot

```python
titanic=sns.load_dataset("titanic")
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
titanic["age"].fillna(titanic["age"].mean(),inplace=True)
```

# plotting a dist plot

```python
sns.distplot(titanic["age"])
```

# plotting only the histogram

```python
sns.distplot(titanic["age"],kde=False)
```

```python
sns.distplot(titanic["age"],bins=25,kde=False)
```

# plotting only the kde

```python
sns.distplot(titanic["age"],hist=False)
```

# plotting the reg plot

```python
sns.distplot(titanic["age"],hist=False,kde=False,rug=True)
```

```python
#how to use the  distplot
```

```python
sns.distplot(titanic[titanic["survived"]==1]["age"])
```

```python
sns.distplot(titanic[titanic["survived"]==1]["age"])
sns.distplot(titanic[titanic["survived"]==0]["age"])
```
