```python
def plot_categorical_variable(df, column_name):
  counts = df[column_name].value_counts()
  ax = counts.plot.bar()
  ax.set_xlabel(column_name)
  ax.set_ylabel('Count')
  ax.set_title(f'{column_name} counts')
```

```python
import matplotlib.pyplot as plt

def plot_line(df, col_name):
    # Create a figure and an axis
    fig, ax = plt.subplots()
    
    # Plot the data
    ax.plot(df[col_name])
    
    # Add a title and axis labels
    ax.set_title("Line plot of " + col_name)
    ax.set_xlabel("Index")
    ax.set_ylabel(col_name)
    
    # Show the plot
    plt.show()

# Select only numerical columns
numerical_cols = df.select_dtypes(include=['float', 'int']).columns

# Iterate over the numerical columns and plot a line plot for each
for col in numerical_cols:
    plot_line(df, col)
```

```python
# with the following function we can select highly correlated features
# it will remove the first feature that is correlated with anything other feature

def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```
